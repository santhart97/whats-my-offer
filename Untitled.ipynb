# Models

- First step is to split the data for Train and Test.
- My aim is to predict the *Price* column with the other variables of *Year*, *Engine Size*, *Brand*, *Fuel Type*, *Transmission*, and *Mileage*.
- When I separate the test and train I take *Price* as the target.
- I will separate the data and take 80% to train and 20% to test.

```python
X = Cars_dum.drop(columns=["Price"])
y = Cars_dum["Price"]


X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
ordered_predictors = X.columns
```

#### Check Models

```python
# Function to run the models and find their scores/
def run_model(X_train, y_train, model, X_test, y_test):
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)
    
    r2_train=r2_score(y_train, y_pred_train)
    r2_test=r2_score(y_test, y_pred_test)
    rmse_train=(mean_squared_error(y_train, y_pred_train))**0.5
    rmse_test=(mean_squared_error(y_test, y_pred_test))**0.5
    
    return f"Train, r2_score: {r2_train}", f"Test, r2_score: {r2_test}", f"Train rmse: {rmse_train}", f"Test, rmse: {rmse_test}"
```

The purpose of the function above is to apply it to every model and receive the r2 score and the mean squared error for the Train and Test data.

## Linear Regression

```python
lr = LinearRegression()
run_model(X_train, y_train, lr, X_test, y_test)
```


- Around 76% of the data fit the linear regression model.
- Train and test data are very similar so there is not much overfitting.

###### Cross validation

```python
scores_lr_r2 = cross_val_score(lr, X_train, y_train, cv=3, scoring="r2")
scores_lr_r2
```

```python
scores_lr_rmse = cross_val_score(lr, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores_lr_rmse
```

### Kneighbors Regression

```python
KN = KNeighborsRegressor(n_neighbors=2)
run_model(X_train, y_train, KN, X_test, y_test)
```

- The scores are of this model are lower than those of Linear Regression.
- Tmrain Score is greater than Test Score, suggesting overfitting.

###### Cross validation

```python
scores_KN_rmse = cross_val_score(KN, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores_KN_rmse
```

```python
scores_KN_r2 = cross_val_score(KN, X_train, y_train, cv=3, scoring="r2")
scores_KN_r2
```

### Random Forest Regression

```python
rand_reg = RandomForestRegressor()
run_model(X_train, y_train, rand_reg, X_test, y_test)
```

- The r2 score is the highest of the three models, however there still is some overfitting.

###### Cross validation

```python
scores_rr_rmse = cross_val_score(rand_reg, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores_rr_rmse
```

```python
scores_rr_r2 = cross_val_score(rand_reg, X_train, y_train, cv=3, scoring="r2")
scores_rr_r2
```

### Decision Tree Regression

```python
DT = DecisionTreeRegressor()
run_model(X_train, y_train, DT, X_test, y_test)
```

- This model returns the highest score but there still is some overfitting (R2 Train > R2 Test)

###### Cross validation

```python
scores_DT_r2 = cross_val_score(DT, X_train, y_train, cv=3, scoring="r2")
scores_DT_r2
```

```python
scores_DT_rmse = cross_val_score(DT, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores_DT_rmse
```

### Gradient Boosting Regression

```python
gbr = GradientBoostingRegressor()
run_model(X_train, y_train, gbr, X_test, y_test)
```

- This model returns a high score and no overfitting.

##### Cross validation

```python
scores_gbr_r2 = cross_val_score(gbr, X_train, y_train, cv=3, scoring="r2")
scores_gbr_r2
```

```python
scores_gbr_rmse = cross_val_score(gbr, X_train, y_train, cv=3, scoring="neg_root_mean_squared_error")
scores_gbr_rmse
```

All the scores obtained from cross validation are similar to the scores we obtained from our models meaning that our metrics are **not** random.
